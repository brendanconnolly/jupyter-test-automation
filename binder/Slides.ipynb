## State,  Output & Execution Flow 
Inside of a notebook data can be shared across cells



When you have multiple code cells variables and functions declared and used in one cell are able to be used in other cells

in this first cell we have 2 variables for different greetings and a function that that joins them into a phrase.

we can run this first cell. Then inside this second cell we can execute it.

you can see that output of the function is rendered below the cell

this is because in many ways jupyter notebooks act like a REPL (Read Evaluate Print loop) that you may be used to either on the console or inside your IDE.

if we change the contents to use a setTimeout to delay the execution of the function

setTimeout(greetingFunction(), 2000); 

=>  when you run this cell the output is the return value of that function, and we wont see the greeting

now if we change the cell 
setTimeout(()=>{console.log(greetingFunction())}, 2000); => now if we add a console log statement to be run after the time out what do you think will happen?

lets run it and see, as soon as the cell is executed the output is rendered, and when the timeout function runs you can see the console greeting

and if you didn't want to display that initial function output, you can add another statement to the cell
    
    `Patiently waiting`
or null/ empty string to display nothing. This might seem a little interesting maybe odd but its actually a common practice for languages that use implicit returns on functions.

## Jupyter Architecture
![jupyter architecture](https://jupyter.readthedocs.io/en/latest/_images/notebook_components.png) 


  
  > *Source:[Jupyter.org](https://jupyter.readthedocs.io/en/latest/projects/architecture/content-architecture.html)*

Users interact with the notebooks via a web browser. The web UI connects to the notebook server which in turn handles working with individual notebook files. 

Language processing is delegated to what is  referred to as a kernel. The Notebook server uses a message queue to delegate lanuage processing to the specific languages kernel. 

The python kernel is included and required, but there are also many community supported language kernels.

```javascript
const axios = require('axios').default;

const roomsUrl ="https://automationintesting.online/room/";
let response = await axios.get(roomsUrl);
let roomResponsedata= response.data;
//roomResponsedata;
```

```javascript
const tslab = require("tslab");

//get the room image using response from initial request
let firstRoomImageData= await axios.get(roomResponsedata.rooms[0].image,{ responseType: 'arraybuffer' })

// tslab is the package for the js kernal with helpers for displaying non-text data
tslab.display.jpeg(new Uint8Array(firstRoomImageData.data));
```

## Using Selenium

Ok sure we can do api calls, but when you see automation in the title I bet you wanted to see selenium and some web automation. 

this code is using webdriverIO to
- launch chrome
- navigate to google
- enter a search string 
- then take a screenshot.


```javascript
const { remote } = require('webdriverio');

let displayScreenShot=(base64String)=>{return `<img style='width:800px;height:600px;' src="data:image/png;base64, ${base64String}" />`}


//configure webdriverIO to use headless chrome through a selenium grid
const browser = await remote({
        logLevel: 'error',
        hostname:'hub',
        path: '/wd/hub',
       
        capabilities: {
            browserName: 'chrome',
            'goog:chromeOptions': { args: ['--headless', '--disable-gpu']}
        }
    });
```

```javascript
const tslab = require("tslab");

//goto google and search then take screenshot
let theUrlToLaunch=`https://google.com`

await browser.url(theUrlToLaunch);
await browser.keys(`Jupyter Notebooks`);
let screenshot= await browser.takeScreenshot();

//display the screenshot we generated 
tslab.display.html(displayScreenShot(screenshot));
```

```javascript
//close the browser
await browser.closeWindow();
```

## Common Components of Automation Frameworks

- Harness
- Driver
- Content
- Information / Data



I don't want to get too caught up in definitions, but to help put Jupyter in context lets briefly talk about 
frameworks and some of theircommon components 

When You Say Framework... I'm not always sure we mean the same things. 

In general terms a framework calls your code, whereas a library has other code call it

There is also a expectation that a framework has multiple consumers different people or things being built with it

- Harness

The core, or the engine of test automation frameworks are built around are unit testing frameworks

this is the code, that calls your code, its what that enables your test code to be run

- Driver

the driver is the code thats used to drive your application. Selenium WebDriver is a driver for browsers
So when we wrap selenium code creating page objects, 
I like to think of it as creating a driver for the UI of the application being tested.
When writing similar type code for APIs I think of it as a driver for that specific application

- Content

Content is the tests, its the s

- Information/Data is input data and output
